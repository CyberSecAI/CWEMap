# Overview

1. Read the LLM Consensus of Clauge and Gemini file
2. Read the ChatGPT4o batch file
3. Merge
4. Plot Consensus

```python
import os
import json
import csv
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
llm_consensus_file_path_sorted = './data_out/llm_consensus_sorted.csv'
chatgpt_batch_file_path = './data_out/chatgpt_batch.csv'
all_llm_consensus_file_path_sorted = './data_out/all_llm_consensus_sorted.csv'
```

```python
df = pd.read_csv(llm_consensus_file_path_sorted)
df

```

```python
df_gp4o = pd.read_csv(chatgpt_batch_file_path)
df_gp4o
```

```python
merged_df = df_gp4o.merge(
    df,
    left_on=['cve_id', 'CWE_ID'],
    right_on=['CVE ID', 'Assigned CWE ID'],
    how='left'
)
merged_df
```

```python
# drop some columns
# move these columns to the end gpt-4o_Agreegpt-4o_Rationalegpt-4o_ConfidenceContainer

merged_df = merged_df.drop(columns=['CVE_Description', 'cna_providerMetadata_orgId', 'cna_providerMetadata_shortName', 'cna_providerMetadata_dateUpdated'])

# List of columns to move to the end
columns_to_move = ['gpt-4o_Agree', 'gpt-4o_Rationale', 'gpt-4o_Confidence']

# Get all column names
all_columns = merged_df.columns.tolist()

# Remove the columns to be moved from the list
for col in columns_to_move:
    if col in all_columns:
        all_columns.remove(col)

# Add the columns to be moved at the end of the list
all_columns.extend(columns_to_move)

# Reorder the DataFrame
merged_df = merged_df[all_columns]

merged_df
```

```python
merged_df.to_csv(all_llm_consensus_file_path_sorted, index=False, quoting=csv.QUOTE_ALL, escapechar='\\') # safe CSV
```

```python
from plotly.io import write_image
def create_parcat_diagram(df, filename='./images/ai_agreement_parcat.png'):
    # Create a function to determine the color based on the agreement pattern
    def get_color(row):
        yes_count = sum(row == 'Yes')
        if yes_count == 3:
            return 0  # Green for 3 Yes
        elif yes_count == 0:
            return 3  # Red for 3 No
        elif yes_count == 2:
            return 1  # Yellow for 2 Yes and 1 No
        else:
            return 2  # Orange for 1 Yes and 2 No

    # Apply the color mapping function
    df['color'] = df.apply(get_color, axis=1)
    
    # Create the parallel categories diagram
    fig = go.Figure(go.Parcats(
        dimensions=[
            {'label': 'Claude', 'values': df['claude_Agreement'], 'categoryorder': 'array', 'categoryarray': ['Yes', 'No']},
            {'label': 'Gemini', 'values': df['gemini_Agreement'], 'categoryorder': 'array', 'categoryarray': ['Yes', 'No']},
            {'label': 'GPT-4', 'values': df['gpt-4o_Agree'], 'categoryorder': 'array', 'categoryarray': ['Yes', 'No']}
        ],
        line={'color': df['color'],
              'colorscale': [[0, 'green'], [0.33, 'yellow'], [0.67, 'orange'], [1, 'red']]},
        hoveron='color',
        hoverinfo='count+probability',
        labelfont={'size': 16, 'family': 'Arial'},
        tickfont={'size': 14, 'family': 'Arial'},
        arrangement='freeform',
    ))

    # Update the layout
    fig.update_layout(
        title={'text': "AI Agreement: Claude Sonnet 3.5 vs Gemini 1.5 Pro vs Chat GPT-4o",
               'font': {'size': 24}},
        height=600,
        width=1000,
    )

    # Save the figure to a file
    write_image(fig, filename)
    print(f"Plot saved to {filename}")


    # Show the figure
    fig.show()

# Create the parallel categories diagram
create_parcat_diagram(merged_df)

```
